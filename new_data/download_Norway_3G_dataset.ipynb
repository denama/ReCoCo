## Downloading Norway 3G dataset and putting it directly into json files

http://skuld.cs.umass.edu/traces/mmsys/2013/pathbandwidth/

```python
import requests
from bs4 import BeautifulSoup
import io
import pandas as pd
import os
import json
```

### Read page content

```python
URL = "http://skuld.cs.umass.edu/traces/mmsys/2013/pathbandwidth/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
# results = soup.find(id="ResultsContainer")
# print(results.prettify())
print(soup)
```

### Take out all links with data and give them names, save them in d_name_link

```python
d_name_link = {}
links = []
names = []
for href in soup.findAll("a"):
    print(href)
    if "LOGS" in str(href):
        additional_url = href["href"]
        url = os.path.join(URL, additional_url)
        print("Url in outer loop: ", url)
        inner_page = requests.get(url)
        inner_soup = BeautifulSoup(inner_page.content, "html.parser")
        for inner_href in inner_soup.findAll("a"):
#             print(type(str(inner_href)))
            if "report" in str(inner_href):
#                 print(type(inner_href["href"]))
                inner_url = os.path.join(url, inner_href["href"])
                name = inner_url.split("/./")[1].split(".")[0] + "_" + inner_href["href"].split(".")[1] 
                print("Name", name)
#                 print("Just inner href: ", inner_href["href"])
                print("Inner url: ", inner_url)
                links.append(inner_url)
                names.append(name)
                d_name_link[name] = inner_url
```

```python
d_name_link
```

### Go through the links one by one, download the data and put it in a json file

```python
save_folder = "../new_data/Norway_3G_data_json/"

for name, link in d_name_link.items():
    s = requests.get(link).content
    df = pd.read_csv(io.StringIO(s.decode('utf-8')), sep=" ", header=None)
    
    save_file = name + ".json"
#     print(save_file)
    
    df.columns = ["epoch_ms", "time", "lat", "long", "bytes", "duration"]
    #calculate capacity in kbps (bytes / duration in ms)
    df["capacity"] = df["bytes"] / df["duration"]
    #convert time to seconds
    df["time"] = df["time"]/1000

    d_final = {
        "type": "video",
        "downlink": {},
        "uplink": {
            "trace_pattern": df[["time", "duration", "capacity"]].to_dict(orient="records")
        },
    }
    
#     print(d_final)
    print("Saving to: ", os.path.join(save_folder,save_file))
    with open(os.path.join(save_folder,save_file), "w") as log_save:
        json.dump(d_final, log_save)
```
